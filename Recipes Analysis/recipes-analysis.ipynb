```python
# read dataset
df = pd.read_csv(r"assets/data/recipes.csv")
print(f'Training Dataset has been read successfully.\nRows : {df.shape[0]}\nColumns : {df.shape[1]}')
```

Some values are counted twices because of string case.

```python
df['country'] = df['country'].str.lower()
df['country'].unique(), df['country'].nunique()
```

```python
cuisine_names = {"austria" : "austrian", "belgium" : "belgian", "china" : "chinese", "canada" : "canadian",
"netherlands" : "dutch", "france" : "french", "germany" : "german", "india" : "indian",
"indonesia" : "indonesian", "iran" : "iranian", "italy" : "italian", "japan" : "japanese",
"israel" : "israeli", "korea" : "korean", "lebanon" : "lebanese", "malaysia" : "malaysian",
"mexico" : "mexican", "pakistan" : "pakistani", "philippines" : "philippine", "scandinavia" : "scandinavian",
"spain" : "spanish_portuguese", "portugal" : "spanish_portuguese", "switzerland" : "swiss", "thailand" : "thai",
"turkey" : "turkish", "vietnam" : "vietnamese", "uk-and-ireland" : "uk-and-irish", "irish" : "uk-and-irish"}

# renaming the cuisine_names
for key, value in cuisine_names.items():
    df.loc[df["country"] == key, "country"] = value

column_names = df.columns.values
column_names[0] = 'cuisine'
df.columns = column_names

df.head()
```

```python
df['cuisine'].nunique()
```

```python
recipes_counts = df["cuisine"].value_counts()

cuisines_indices = recipes_counts > 50

cuisines_to_keep = list(np.array(recipes_counts.index.values)[np.array(cuisines_indices)])
print(cuisines_to_keep)
```

```python
rows_before = df.shape[0]
print("Number of rows of original dataframe is {}.".format(rows_before))

recipes = df.loc[df['cuisine'].isin(cuisines_to_keep)]

rows_after = recipes.shape[0]
print("Number of rows of processed dataframe is {}.".format(rows_after))

print("{} rows removed!".format(rows_before - rows_after))
```

```python
# %timeit recipes.replace(to_replace="Yes", value=1)
# %timeit recipes.replace(to_replace="No", value=0)
```

```python
recipes = recipes.replace(to_replace="Yes", value=1)
recipes = recipes.replace(to_replace="No", value=0)
```

```python
recipes.head()
```

```python
ing = recipes.iloc[:, 1:].sum(axis=0)

ingredient = pd.Series(ing.index.values, index = np.arange(len(ing)))

count = pd.Series(list(ing), index = np.arange(len(ing)))
```

```python
ing_df = pd.DataFrame(dict(ingredient = ingredient, count = count))
ing_df = ing_df[["ingredient", "count"]]
```

```python
ing_df.sort_values(["count"], ascending=False, inplace=True)
ing_df.reset_index(inplace=True, drop=True)

print(ing_df)
```

```python
cuisines = recipes.groupby("cuisine").mean()
cuisines.head()
```

Lets make final move to draw out the top 4 most used ingredient from each country

```python
# number of top ingredients to print
num_ingredients = 4 

# a function that prints the top ingredients for each cuisine
def print_top_ingredients(row):
    print(row.name.upper())
    row_sorted = row.sort_values(ascending=False)*100
    top_ingredients = list(row_sorted.index.values)[0:num_ingredients]
    row_sorted = list(row_sorted)[0:num_ingredients]

    for ind, ingredient in enumerate(top_ingredients):
        print("%s (%d%%)" % (ingredient, row_sorted[ind]), end=', ')
    print("\n")

# apply function to cuisines dataframe
create_cuisines_profiles = cuisines.apply(print_top_ingredients, axis=1)
```
